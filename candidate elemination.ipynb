```python
dat={
    "sky":['sunny','sunny','rainy','sunny'],
    "airTemp":['warm','warm','cold','warm'],
    "humidity":['normal','high','high','high'],
    "wind":['strong','strong','strong','strong'],
    "water":['warm','warm','warm','cold'],
    "forecast":['same','same','change','change'],
    "enjoysport":['yes','yes','no','yes']
}
```

```python
import pandas as pd
```

```python
dataset=pd.DataFrame(data)
dataset
```

```python
import numpy as np
import pandas as pd

data = pd.DataFrame(dat)
concepts = np.array(data.iloc[:,:-1])
print("\nintialization of specific_h and general_h")
specific_h= concepts[0].copy()
print("\nspecific boundary : ",specific_h)
target= np.array(data.iloc[:,-1])
print("\nTarget values are : ",target)

def learn(concepts,target):
    specific_h=concepts[0].copy()
    print("\nspecific Boundry",specific_h)
    genral_h=[["?" for i in range(len(specific_h))] for i in range(len(specific_h))]
    print("genral",genral_h)
    
    for i,h in enumerate(concepts):
        print('\nInstance',i+1,'is',h)
        if target[i]=='yes':
            print("Instance is positive")
            for x in range(len(specific_h)):
                if h[x]!= specific_h[x]:
                    specific_h[x]='?'
                    genral_h[x][x]='?'
        if target[i]== "no":
            print("Instance is Negative")
            for x in range(len(specific_h)):
                if h[x]!= specific_h[x]:
                    genral_h[x][x] = specific_h[x]
                else:
                    genral_h[x][x]='?'
        print("Specific Boundary after",i+1, "Instance is ", specific_h)
        print("Specific Boundary after",i+1, "Instance is ", genral_h)
        print("/n")
    
    indices = [i for i, val in enumerate(genral_h) if val == ['?','?','?','?','?','?']]
    for i in indices:
        genral_h.remove(['?','?','?','?','?','?'])
    return specific_h, genral_h
               
s_final,g_final = learn(concepts, target)

print("Final Specific_h: ", s_final, sep="/n")
print("Final General_h: ", g_final, sep="/n")

    
    
```
